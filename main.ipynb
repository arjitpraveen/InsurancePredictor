# Definition and Working Principle

We will build a linear regression model to help predict insurance charges based off of various demographic and health-related factors from the `insurance.csv` dataset.

**Linear regression** is a statistical analysis technique used to model the relationship between an independent variable(s) and a dependent variable by fitting a regression line. The model estimates parameters - weights and bias - that quantify the relationship. Furthermore, we shall calculate $R^2$ to quantify how much of the variation in the observed data can be explained by the regression model.

## Hypothesis Representation

Let the independent and dependent variables be $\text{x}_i$ and $\text{y}_i$ respectively, where $\text{i}$ is the feature value. Given a training set, we will model a hypothesis function $\text{h}$ to estimate $\text{y}$ based on $\text{x}$. The hypothesis function for *simple linear regression* is given by,

$$\text{y}_i = \text{h}_\theta (\text{x}_i) = {\theta}_0 + {\theta}_1 \text{x}_i$$

Where ${\theta}_0$ and ${\theta}_1$ are the model parameters. 

For *multiple linear regression*, where more than 1 independent variable/feature exists, we will use $\text{x}_{ij}$ and $\text{y}_i$ to denote the independent variables and the dependent variable respectively. Given $\text{m}$ training samples and $\text{n}$ features such that $\text{ j} = 1, 2, 3, \dots, n$,  the hypothesis function $\text{h}$ is given by,

$$\text{h}_\theta (\text{x}_i) = {\theta}_0 + {\theta}_1 \text{x}_{i1} + {\theta}_2 \text{x}_{i2} + \dots + {\theta}_j \text{x}_{ij} + \dots + {\theta}_n \text{x}_{mn}$$

where $\text{x}_{ij}$ is the $i^{th}$ training sample of the $j^{th}$ feature, and $\theta_0, \theta_1, \dots, \theta_i, \dots, \theta_n$ are the model parameters.

```python
import pandas as pd
```

```python
insurance = pd.read_csv('data/insurance.csv')
insurance.head()
```

```python
print("rows, columns:", insurance.shape)
print("columns:", insurance.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
insurance_copy = insurance.copy()
encoded_df = insurance_copy

labelencoder = LabelEncoder()
categorical_columns = ['sex', 'region', 'smoker']

for column in categorical_columns:

    columns_data = encoded_df[[column]]
    encoded_df[column + '_encoded'] = labelencoder.fit_transform(columns_data[column])
    encoded_df.drop(columns=[column], inplace=True)

X_encoded = encoded_df.drop("charges", axis=1)
y = encoded_df['charges']
```
